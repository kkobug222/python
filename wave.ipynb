```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8  # gravitational acceleration (m/s^2)

#1: Wave speed comparison
wavelength = 100  # wavelength for deep water wave (m)
depth_range = np.linspace(1, 100, 500)  # depth values (m)


speed_shallow = np.sqrt(g * depth_range)  # shallow water
speed_deep = np.sqrt((g * wavelength) / (2 * np.pi))  # deep water wave
speed_deep_array = np.full_like(depth_range, speed_deep)

# Plot wave speed comparison
plt.figure(figsize=(10, 5))
plt.plot(depth_range, speed_shallow, label='Shallow Water Wave Speed (√gh)', color='blue')
plt.plot(depth_range, speed_deep_array, label='Deep Water Wave Speed (√(gλ/2π))', color='red', linestyle='--')
plt.title("Wave Speed vs. Depth")
plt.xlabel("Water Depth (m)")
plt.ylabel("Wave Speed (m/s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#2: Tsunami simulation (Shoaling effect)

tsunami_wavelength = 200_000  # large wavelength (200 km)
initial_height = 1  # initial wave height (m)

# Depth profile: ocean to shore
depth_profile = np.linspace(5000, 10, 500)  # decreasing depth (m)
tsunami_speed = np.sqrt(g * depth_profile)  # assume tsunami behaves like shallow water wave

# Wave height increases inversely with speed (energy conservation)
wave_heights = initial_height * (tsunami_speed[0] / tsunami_speed)

# Distance axis
distance_axis = np.linspace(0, 500, 500)  # km

# Plot tsunami shoaling
plt.figure(figsize=(10, 5))
plt.plot(distance_axis, wave_heights, color='green')
plt.title("Tsunami Shoaling Effect: Wave Height vs. Distance")
plt.xlabel("Distance from Shore (km)")
plt.ylabel("Relative Wave Height (m)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
